# Denoising the simulated Phantom with different algorithms

```python
import sys
sys.path.append('..')

from model.patch2self import patch2self

from dipy.io.image import load_nifti, save_nifti
import matplotlib.pyplot as plt
import numpy as np
```

#### Load the noisy simulated phantom

```python
data10, affine = load_nifti('DWI62_snr10.nii.gz')
data15, affine = load_nifti('DWI62_snr15.nii.gz')
data20, affine = load_nifti('DWI62_snr20.nii.gz')
data25, affine = load_nifti('DWI62_snr25.nii.gz')
data30, affine = load_nifti('DWI62_snr30.nii.gz')
```

#### Denoise SNR 10 Noisy Phantom with Patch2Self

```python
den_p2s_10 = patch2self(data10)
```

#### Denoise SNR 15 Noisy Phantom with Patch2Self

```python
den_p2s_15 = patch2self(data15)
```

#### Denoise SNR 20 Noisy Phantom with Patch2Self

```python
den_p2s_20 = patch2self(data20)
```

#### Denoise SNR 25 Noisy Phantom with Patch2Self

```python
den_p2s_25 = patch2self(data25)
```

#### Denoise SNR 30 Noisy Phantom with Patch2Self

```python
den_p2s_30 = patch2self(data30)
```

#### Plots of the Simulated Phantom  with Patch2Self

```python
sli = 2
vol = 15

d10 = data10[:, :, sli, vol]
den10 = den_p2s_10[:, :, sli, vol]

d15 = data15[:, :, sli, vol]
den15 = den_p2s_15[:, :, sli, vol]

d20 = data20[:, :, sli, vol]
den20 = den_p2s_20[:, :, sli, vol]

d25 = data25[:, :, sli, vol]
den25 = den_p2s_25[:, :, sli, vol]

d30 = data30[:, :, sli, vol]
den30 = den_p2s_30[:, :, sli, vol]

fig, ax = plt.subplots(2, 5, figsize=(30, 30))
ax[0][0].imshow(d10.T, cmap='gray', origin='lower', interpolation='None')
ax[0][0].set_title('SNR 10')
ax[0][0].set_axis_off()
ax[1][0].imshow(den10.T, cmap='gray', origin='lower', interpolation='None')
ax[1][0].set_title('SNR 10: Patch2Self')
ax[1][0].set_axis_off()
ax[0][1].imshow(d15.T, cmap='gray', origin='lower', interpolation='None')
ax[0][1].set_title('SNR 15')
ax[0][1].set_axis_off()
ax[1][1].imshow(den15.T, cmap='gray', origin='lower', interpolation='None')
ax[1][1].set_title('SNR 15: Patch2Self')
ax[1][1].set_axis_off()
ax[0][2].imshow(d20.T, cmap='gray', origin='lower', interpolation='None')
ax[0][2].set_title('SNR 20')
ax[0][2].set_axis_off()
ax[1][2].imshow(den20.T, cmap='gray', origin='lower', interpolation='None')
ax[1][2].set_title('SNR 20: Patch2Self')
ax[1][2].set_axis_off()
ax[0][3].imshow(d25.T, cmap='gray', origin='lower', interpolation='None')
ax[0][3].set_title('SNR 25')
ax[0][3].set_axis_off()
ax[1][3].imshow(den25.T, cmap='gray', origin='lower', interpolation='None')
ax[1][3].set_title('SNR 25: Patch2Self')
ax[1][3].set_axis_off()
ax[0][4].imshow(d30.T, cmap='gray', origin='lower', interpolation='None')
ax[0][4].set_title('SNR 30')
ax[0][4].set_axis_off()
ax[1][4].imshow(den30.T, cmap='gray', origin='lower', interpolation='None')
ax[1][4].set_title('SNR 30: Patch2Self')
ax[1][4].set_axis_off()
plt.tight_layout()
plt.subplots_adjust(hspace=0, top=0.45)
plt.show()
```

#### Plots of denoising with Marchenko-Pastur PCA and Adaptive Optimized Non-Local Means

```python
# obtained from the MPPCA implementation of Mrtrix (denoised output provided with supplement)
lpca10, affine = load_nifti('den_62_snr10.nii.gz')
lpca15, affine = load_nifti('den_62_snr15.nii.gz')
lpca20, affine = load_nifti('den_62_snr20.nii.gz')
lpca25, affine = load_nifti('den_62_snr25.nii.gz')
lpca30, affine = load_nifti('den_62_snr30.nii.gz')

# obtained from the AONLM implementation avilable at:
# https://sites.google.com/site/pierrickcoupe/softwares/denoising-for-medical-imaging/mri-denoising/mri-denoising-software  
# (denoised output provided with supplement)
nlm10, affine = load_nifti('DWI62_snr10_denoised_AONLM.nii')
nlm15, affine = load_nifti('DWI62_snr15_denoised_AONLM.nii')
nlm20, affine = load_nifti('DWI62_snr20_denoised_AONLM.nii')
nlm25, affine = load_nifti('DWI62_snr25_denoised_AONLM.nii')
nlm30, affine = load_nifti('DWI62_snr30_denoised_AONLM.nii')
```

```python
sli = 2
vol = 15

l10 = lpca10[:, :, sli, vol]
n10 = nlm10[:, :, sli, vol]

l15 = lpca15[:, :, sli, vol]
n15 = nlm15[:, :, sli, vol]

l20 = lpca25[:, :, sli, vol]
n20 = nlm25[:, :, sli, vol]

l25 = lpca25[:, :, sli, vol]
n25 = nlm25[:, :, sli, vol]

l30 = lpca30[:, :, sli, vol]
n30 = nlm30[:, :, sli, vol]

fig, ax = plt.subplots(2, 5, figsize=(30, 30))
ax[0][0].imshow(l10.T, cmap='gray', origin='lower', interpolation='None')
ax[0][0].set_title('SNR 10: MPPCA')
ax[0][0].set_axis_off()
ax[1][0].imshow(n10.T, cmap='gray', origin='lower', interpolation='None')
ax[1][0].set_title('SNR 10: AONLM')
ax[1][0].set_axis_off()
ax[0][1].imshow(l15.T, cmap='gray', origin='lower', interpolation='None')
ax[0][1].set_title('SNR 15: MPPCA')
ax[0][1].set_axis_off()
ax[1][1].imshow(n15.T, cmap='gray', origin='lower', interpolation='None')
ax[1][1].set_title('SNR 15: AONLM')
ax[1][1].set_axis_off()
ax[0][2].imshow(l20.T, cmap='gray', origin='lower', interpolation='None')
ax[0][2].set_title('SNR 20: MPPCA')
ax[0][2].set_axis_off()
ax[1][2].imshow(n20.T, cmap='gray', origin='lower', interpolation='None')
ax[1][2].set_title('SNR 20: AONLM')
ax[1][2].set_axis_off()
ax[0][3].imshow(l25.T, cmap='gray', origin='lower', interpolation='None')
ax[0][3].set_title('SNR 25: MPPCA')
ax[0][3].set_axis_off()
ax[1][3].imshow(n25.T, cmap='gray', origin='lower', interpolation='None')
ax[1][3].set_title('SNR 25: AONLM')
ax[1][3].set_axis_off()
ax[0][4].imshow(l30.T, cmap='gray', origin='lower', interpolation='None')
ax[0][4].set_title('SNR 30: MPPCA')
ax[0][4].set_axis_off()
ax[1][4].imshow(n30.T, cmap='gray', origin='lower', interpolation='None')
ax[1][4].set_title('SNR 30: AONLM')
ax[1][4].set_axis_off()
plt.tight_layout()
plt.subplots_adjust(hspace=0, top=0.45)
plt.show()
```
