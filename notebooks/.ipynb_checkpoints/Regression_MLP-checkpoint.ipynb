#### Import all the necessary packages as well as the module to perform Patch2Self

```python
import sys
sys.path.append("..")

import numpy as np
import time
import matplotlib.pyplot as plt
from model.patch2self import patch2self
from dipy.io.image import save_nifti, load_nifti
from dipy.data import read_stanford_labels
```

#### To showcase the denoising from Patch2Self we will make use of the Stanford HARDI dataset available as a part of DIPY [https://dipy.org]. The 'get_data()' utility is the fetcher that automatically downloads the dataset and loads the Nifti file as follows:

```python
hardi_img, gtab, labels_img = read_stanford_labels()
data = hardi_img.get_data()
affine = np.eye(4)
```

#### Patch2Self via Ordinary Least Squares Regression

```python
t1 = time.time()
denoised_sh = patch2self(data, model='ols')
t2 = time.time()
print('Time Taken: ', t2-t1)
```

- #### Next we will visualize the output of the denoising and plot the residuals using sum of squared differences
- #### We show the output for slice 35 and volume 127

```python
sli = 35
vol = 127
orig1 = data[:, :, sli, vol]

den1 = denoised_sh[:, :, sli, vol]
rms_diff1 = np.sqrt((orig1 - den1) ** 2)

fig, ax = plt.subplots(1, 3, figsize=(30, 30))
ax[0].imshow(orig1.T, cmap='gray', origin='lower', interpolation='None')
ax[0].set_title('Original')
ax[0].set_axis_off()
ax[1].imshow(den1.T, cmap='gray', origin='lower', interpolation='None')
ax[1].set_title('Patch2Self')
ax[1].set_axis_off()
ax[2].imshow(rms_diff1.T, cmap='gray', origin='lower', interpolation='None')
ax[2].set_title('Residual')
ax[2].set_axis_off()
plt.show()
```

#### Patch2Self via Ridge Regression

```python
t1 = time.time()
denoised_sh_ridge = patch2self(data, model='ridge')
t2 = time.time()
print('Time Taken: ', t2-t1)
```

#### Visualizing the outputs as follows:

```python
sli = 35
vol = 127
orig1 = data[:, :, sli, vol]

den1 = denoised_sh_ridge[:, :, sli, vol]
rms_diff1 = np.sqrt((orig1 - den1) ** 2)

fig, ax = plt.subplots(1, 3, figsize=(30, 30))
ax[0].imshow(orig1.T, cmap='gray', origin='lower', interpolation='None')
ax[0].set_title('Original')
ax[0].set_axis_off()
ax[1].imshow(den1.T, cmap='gray', origin='lower', interpolation='None')
ax[1].set_title('Patch2Self')
ax[1].set_axis_off()
ax[2].imshow(rms_diff1.T, cmap='gray', origin='lower', interpolation='None')
ax[2].set_title('Residual')
ax[2].set_axis_off()
plt.show()
```

#### Patch2Self via Lasso Regression

```python
import warnings
warnings.filterwarnings('ignore')

t1 = time.time()
denoised_sh_lasso = patch2self(data, model='lasso')
t2 = time.time()
print('Time Taken: ', t2-t1)
```

#### Visualizing the outputs

```python
sli = 35
vol = 127
orig1 = data[:, :, sli, vol]

den1 = denoised_sh_lasso[:, :, sli, vol]
rms_diff1 = np.sqrt((orig1 - den1) ** 2)

fig, ax = plt.subplots(1, 3, figsize=(30, 30))
ax[0].imshow(orig1.T, cmap='gray', origin='lower', interpolation='None')
ax[0].set_title('Original')
ax[0].set_axis_off()
ax[1].imshow(den1.T, cmap='gray', origin='lower', interpolation='None')
ax[1].set_title('Patch2Self')
ax[1].set_axis_off()
ax[2].imshow(rms_diff1.T, cmap='gray', origin='lower', interpolation='None')
ax[2].set_title('Residual')
ax[2].set_axis_off()
plt.show()
```

#### Computing the self-supervised loss for each of the 160 volumes of the Stanford HARDI dataset for each of the regression models used above:

```python
self_supervised_loss_OLS = []
for i in range(0, data.shape[3]):
    loss = np.mean((data[..., i] - denoised_sh[..., i])**2)
    self_supervised_loss_OLS.append(loss.mean())
    
self_supervised_loss_ridge = []
for i in range(0, data.shape[3]):
    loss = np.mean((data[..., i] - denoised_sh_ridge[..., i])**2)
    self_supervised_loss_ridge.append(loss.mean())
    
self_supervised_loss_lasso = []
for i in range(0, data.shape[3]):
    loss = np.mean((data[..., i] - denoised_sh_lasso[..., i])**2)
    self_supervised_loss_lasso.append(loss.mean())
```

#### Making a dataframe of all losses:

```python
import pandas as pd

df = pd.DataFrame({"Volume":np.arange(160), # 160 because the Stanford HARDI dataset has 160 volumes
                   "OLS":self_supervised_loss_OLS, 
                   "Ridge":self_supervised_loss_ridge,
                   "Lasso":self_supervised_loss_lasso}) 
```

#### Line plots corresponding to each regressor:

```python
%matplotlib inline
df.plot(x="Volume", y=["OLS", "Ridge", "Lasso"], kind="line")
plt.legend(fontsize='xx-large')
```
