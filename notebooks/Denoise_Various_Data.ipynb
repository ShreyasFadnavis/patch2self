```python
import sys
sys.path.append('..')

from model.patch2self import patch2self

from dipy.io.image import save_nifti, load_nifti
from dipy.io import read_bvals_bvecs

import numpy as np
```

#### Load the Sherbrooke 3 Shell Data and the Marchenko Denoised Version

```python
data, _ = load_nifti('../../Desktop/Denoising_Phantom_Experiments/HARDI193.nii.gz')
s3s_mp, _ = load_nifti('../../Desktop/Denoising_Phantom_Experiments/s3sh_mp.nii.gz')
```

#### Denoise the data with Patch2Self

```python
s3s_p2s = patch2self(data)
```

#### Visualize the outputs

```python
import matplotlib.pyplot as plt

sli = 30
vol = 44

orig = data[:, :, sli, vol]

den = s3s_mp[:, :, sli, vol]
rms_diff = np.sqrt((orig - den) ** 2)

fig, ax = plt.subplots(1, 3, figsize=(30, 30))
ax[0].imshow(orig.T, cmap='gray', origin='lower', interpolation='None')
ax[0].set_title('Original Data')
ax[0].set_axis_off()
ax[1].imshow(den.T, cmap='gray', origin='lower', interpolation='None')
ax[1].set_title('Patch2Self')
ax[1].set_axis_off()
ax[2].imshow(rms_diff.T, cmap='gray', origin='lower', interpolation='None')
ax[2].set_title('Residual')
ax[2].set_axis_off()
plt.show()
```

```python
sli = 30
vol = 44

orig = data[:, :, sli, vol]

den = s3s_p2s[:, :, sli, vol]
rms_diff = np.sqrt((orig - den) ** 2)

fig, ax = plt.subplots(1, 3, figsize=(30, 30))
ax[0].imshow(orig.T, cmap='gray', origin='lower', interpolation='None')
ax[0].set_title('Original Data')
ax[0].set_axis_off()
ax[1].imshow(den.T, cmap='gray', origin='lower', interpolation='None')
ax[1].set_title('Patch2Self')
ax[1].set_axis_off()
ax[2].imshow(rms_diff.T, cmap='gray', origin='lower', interpolation='None')
ax[2].set_title('Residual')
ax[2].set_axis_off()
plt.show()
```
